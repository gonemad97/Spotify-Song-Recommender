```python
from sklearn.preprocessing import StandardScaler

features = data_reordered.loc[:, 'energy':'speechiness']
cols_to_standardize = features.columns.tolist()
data_to_standardize = data_reordered[cols_to_standardize]
data_to_standardize
# #data_to_standardize = data_reordered.loc[:, 'energy':'speechiness']
scaler = StandardScaler().fit(data_to_standardize)

# Standardize the columns.
standardized_data = data_reordered.copy()
standardized_columns = scaler.transform(data_to_standardize)
standardized_data[cols_to_standardize] = standardized_columns
standardized_data = standardized_data[cols_to_standardize]

# Fit the model.
from sklearn.cluster import KMeans
model = KMeans(n_clusters=5).fit(standardized_data[cols_to_standardize])

# # Get results.
data_reordered['cluster'] = model.predict(standardized_data)
data_reordered['cluster']
data_reordered
```

```python
#third step of PCA
pca = PCA(n_components=7)
pca.fit(std)

score_pca = pca.transform(std)
score_pca
```

```python
#kmeans_pca impl
kmeans_pca = KMeans(n_clusters=5,init="k-means++")
kmeans_pca.fit(score_pca)
```

```python
#getting the labels in the new PCA Kmeans df
segm_kmeans = pd.concat([data_reordered.reset_index(drop=True)],axis=1)
segm_kmeans.columns.values[-7:] = ["Component 1","Component 2","Component 3","Component 4","Component 5","Component 6","Component 7"]
segm_kmeans["Segment KMeans PCA"] = kmeans_pca.labels_
segm_kmeans.head()
segm_kmeans["Segment"] = segm_kmeans["Segment KMeans PCA"].map({0:"Cluster 1",1:"Cluster 2",2:"Cluster 3",3:"Cluster 4",4:"Cluster 5"})
```

```python
#visualization PCA Kmeans result
%matplotlib inline
import seaborn as sns

x_axis = segm_kmeans["Component 7"]
y_axis = segm_kmeans["Component 4"]
plt.figure(figsize=(15,15))
sns.scatterplot(x_axis,y_axis,hue = segm_kmeans["Segment"],palette = ['g','r','c','m','y'])
plt.title("Clusters by PCA Components")
plt.show()
```
